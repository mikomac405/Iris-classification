[*Klasyfiacja dla list, Klasyfikacja dla DataFrame, Obliczanie wartości obiektów do klasyfikacji*]

```
# importy
import pandas as pd
import random as rd
```

```
class SoftSet:
    @staticmethod
    def classify_list(x,y,test):
        #obliczenie składowych
        values = []
        for i in range(len(y)):
            score = 0
            for param in test.keys():
                score += y[i][param]*test[param]
            values.append(score)

        #znalezienie maksimum
        max_val = max(values)
        print(max_val)
        results = []
        for i in range(len(x)):
            if values[i] == max_val:
                results.append(x[i])

        #zwrócenie wyników
        return results

    @staticmethod
    def classify_dataframe(x,test):
        values = []
        for i in range(len(x.index)):
            score = 0
            for param in x.columns:
                score += x.iloc[i][param]*test[param]
            values.append(score)

        #znalezienie maksimum
        max_val = max(values)
        results = []
        for i in range(x.count()[0]):
            if values[i] == max_val:
                results.append(x.index[i])

        #zwrócenie wyników
        return results


    
    @staticmethod
    def calculate_values(x):
        # sepal.length, sepal.width, petal.length, petal.width, variety
        # _Setosa, _Virginica, _Versicolor
        iris_mean = x.mean()
        s_mean = x.where(x['variety'] == 'Setosa').mean()
        vi_mean = x.where(x['variety'] == 'Virginica').mean()
        ve_mean = x.where(x['variety'] == 'Versicolor').mean()
    

        s_values = {'sepal.length':int(s_mean[0]>iris_mean[0]),
                    'sepal.width':int(s_mean[1]>iris_mean[1]),
                    'petal.length':int(s_mean[2]>iris_mean[2]),
                    'petal.width':int(s_mean[3]>iris_mean[3])}

        vi_values = {'sepal.length':int(vi_mean[0]>iris_mean[0]),
                    'sepal.width':int(vi_mean[1]>iris_mean[1]),
                    'petal.length':int(vi_mean[2]>iris_mean[2]),
                    'petal.width':int(vi_mean[3]>iris_mean[3])}

        ve_values = {'sepal.length':int(ve_mean[0]>iris_mean[0]),
                    'sepal.width':int(ve_mean[1]>iris_mean[1]),
                    'petal.length':int(ve_mean[2]>iris_mean[2]),
                    'petal.width':int(ve_mean[3]>iris_mean[3])}



        result = pd.DataFrame(index=['Setosa','Virginica','Versicolor'],data=[s_values,vi_values,ve_values])

        return result

                
        
```

# 3. <span style="color:green"> Przykład użycia zbioru miękkiego dla irysów:
[*Operowanie na DataFrame*]

```
class DataProcessing:
    @staticmethod
    def shuffle(x):
        for i in range(len(x)-1,0,-1):
            j = rd.randint(0,i)
            x.iloc[i], x.iloc[j] = x.iloc[j], x.iloc[i]
        return x

    @staticmethod
    def split_set(x):
        k = int(len(x)*0.7)
        train = x[:k]
        val = x[k:]
        return train, val
    
    @staticmethod
    def normalize(x):
        val = x.select_dtypes(exclude="object")
        columns = val.columns.tolist()
        for c in columns:
            current_set = x.loc[:, c]
            maximum, minimum = max(current_set), min(current_set)
            for r in range(0,len(x),1):
                value = (x.at[r, c] - minimum) / (maximum - minimum)
                x.at[r, c] = value
```

```
iris = pd.read_csv('iris.csv')
iris.head(50)
```

```
DataProcessing.normalize(iris)
irisTrain, irisVal = DataProcessing.split_set(DataProcessing.shuffle(iris))
```

```
irisTrain = SoftSet.calculate_values(irisTrain)
irisVal = irisVal.reset_index()

correct = 0
for i in range(irisVal.count()[0]):
    print("========================= TEST",i,"=========================")
    guess = SoftSet.classify_dataframe(irisTrain,irisVal.iloc[i])
    anwser = [irisVal.iloc[i]['variety']]
    print('Guess:', guess,' | Anwser:', anwser)
    print()
    if guess == anwser:
        correct += 1

print("Accuracy:", correct/irisVal.count()[0])
```
